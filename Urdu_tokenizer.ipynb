```python
def my_split(String, separator): #to split the string but keep the punctuation
    
    #dummy character to replace separator
    temp = chr(127+1) 
    
    concat = separator + temp
    String = String.replace(separator,concat)
    return String.split(temp)
```

```python
conj_list = ['جنہیں', 'جس', 'جن', 'جو', 'اور', 'اگر', 'اگرچہ', 'لیکن', 'مگر', 'پر', 'یا', 'تاہم', 'کہ', 'کر',
                  'تو', 'گے', 'گی']
```

```python
def sentence_split(sent):
    
    sent = sent.split()
    sent_tokens = []
    flag = False
    sent_stopper = ["۔", "،","?"] #if already present at the end of the senetence
    length = len(sent)
    new = ""
    for i in range(0,len(sent)):
        if flag:
            flag = False
            continue
        #Rule: the last word of the sentence must be a new line word like (hai,hain) and 
        #the next word to the sentence ending word must not be a urud conjunction
        if sent[i] in end_word_list and sent[i + 1] not in conj_list:
            
            #check for conjunctions in the dictionary
            print(sent[i])
            if sent[i + 1] in conjunctions[sent[i]]:
                continue
            
            #print(sent[i])
            if (i == length-1):
                #print(new)
                sent_tokens.append(new.strip())
            
            #separating if the sentence ends in '-' or ',' or "?"
            if sent[i + 1] in sent_stopper:
                new += " " + sent[i] + " " + sent[i + 1]
                sent_tokens.append(new.strip())
                new=""
                #print("called1")
                flag = True #continue if the sentence ends in ''
                continue
                
            else:
                new += " " 
                new += sent[i]
                #print("called2")
                sent_tokens.append(new.strip())
                new=""
                continue

        else:
            new += " " 
            new += sent[i]
            
        if i == length-1:
            sent_tokens.append(new.strip())
    
    return sent_tokens
```
